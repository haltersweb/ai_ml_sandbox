# Model Representation


## Linear Regression with One Variable

### Goal

* Implement the model $f_{w,b}$ for linear regression with one variable.

### Notation

| General Notation | Description | Python (if applicable) |
|:---|:---|---:|
| $a$ | scalar (non bold) ||
| $\mathbf{a}$ | vector (**bold**) ||
| **Regression** |||
| $\mathbf{x}$ | Training Example feature values | `x_train` |   
| $\mathbf{y}$ | Training Example  targets | `y_train` |
| $x^{(i)}$, $y^{(i)}$ | $i_{th}$Training Example | `x_i`, `y_i`|
| m | Number of training examples | `m`|
| $w$ | parameter: weight | `w` |
| $b$ | parameter: bias | `b` |     
| $f_{w,b}(x^{(i)})$ | The result of the model evaluation at $x^{(i)}$ parameterized by $w,b$:<br />$f_{w,b}(x^{(i)}) = wx^{(i)}+b$ | `f_wb` |

### Packages

* NumPy
* MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('deeplearning.mplstyle')
```

### Problem Statement

Fit a linear regression model through two data points in order to predict price for other houses.

| Size (1000 sqft) | Price (1000s of dollars) |
|---:|---:|
| 1.0	| 300|
| 2.0	| 50|

### The Data

#### training data as vectors

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train shape = {x_train.shape}") #tuple of w,h (no h if vector)
```

#### number of training examples

```python
m = x_train.shape[0]
print(f"number of training examples = {m}")
```

#### $i_{th}$ training example

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

#### scatter plot the data
Use **matplotlib**'s `scatter()` function:
* mark: x
* color: red
* title and axis labels

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Housing Prices')
plt.xlabel('Sq Ft (in 1000s)')
plt.ylabel('Price (in 1000 dollars)')
plt.show()
```

### The Model

The **model function** for **linear regression** is:
$$
f_{w,b}(x^{(i)}) = wx^{(i)} + b
$$




### Predicting the Model

Find the linear regression model `f_wb` = $f_{w,b}(x^{(i)}) = wx^{(i)} + b$

#### set parameters

* **x** `x_train
* weight `w = 100`
* bias `b = 100`


```python
#we'll be using np.zeros()
foo = np.zeros(5)
print(foo)
```

```python
w = 100 #weight
b = 100 #bias
print(f"w: {w}")
print(f"b: {b}")
```

#### prediction function

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
        x (ndarray (m,))    : Data, m examples
        w,b (scalar)        : model parameters
    Returns:
        f_wb (ndarray (m,)) : model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

#### get results of $f_{(w,b)}x$

```python
#prediction results
f_wb = compute_model_output(x_train, w, b)
print(f"predicted model is {f_wb}")
```

#### plot the prediction results

Plot the prediction `f_wb` relative to training data

```python
plt.plot(x_train, f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.scatter(x_train, y_train, c='r', marker='x', label='Actual Values')
plt.title('Housing Prices')
plt.xlabel('Sq Ft (in 1000s)')
plt.ylabel('Price (in 1000 dollars)')
plt.legend()
plt.show()
```

#### make a better guess

Come up with better `w` and `b`

```python
w = 200
b = 100
f_wb = compute_model_output(x_train, w, b)
```

```python
plt.plot(x_train, f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.scatter(x_train, y_train, c='r', marker='x', label='Actual Values')
plt.title('Housing Prices')
plt.xlabel('Sq Ft (in 1000s)')
plt.ylabel('Price (in 1000 dollars)')
plt.legend()
plt.show()
```
